### [Image classification](https://huggingface.co/tasks/image-classification)

e.g. https://huggingface.co/google/vit-base-patch16-224 \
How to use

```python
!pip install transformers

# 트랜스포머 살치 
```

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests

url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
image = Image.open(requests.get(url, stream=True).raw)
# 이미지 불러온다, 

feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
# pretrained 행렬이다. 
# 피처 익스트렌트를 할 수있는 모델이다. 클래시파이가 가능하다. 
# 변수명을 보고 추측하자. 클래시파이하는 모델이다. 
# 피처란? 흑백사진이 있으면 밝으면 1, 어두우면 0 이런식 숫자 정보이다. 
# 만약 칼라면 그라한 것들을 rgb의 3층 정보가 있다는 것. 그것을 다쓰기에는 너무 많다. 
# 1장짜리라도 숫자들 속에서 정보들을 뽑아내는 것


inputs = feature_extractor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
# model predicts one of the 1000 ImageNet classes
# 정보라는 것이 무한개이다. 그것을 피처형태로 뽑으면 그 어떤 오브젝트를 정의하고 기술하는 핵심
# 리미트된 넘버들의 피처- 특징들을 많은 수로부터 추출가능하다. 뽑아낸다. 
# 오리지날 넘버보다는 뽑아낸 것이 더욱 작을 것이다. 그 숫자들을 압축된 정보로 모델을 입력
# 클래시파이된 정보를 잡는다. 1000개의 숫자를 다 더하면 1이 된다. 
# 확률이 가장 큰 것을 찾는다. 1개만 찾는다? 그제일 확률 높은 것을 찾는다. 

predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])

# 따닥다다 나온다. 대충 읽어보자. 
# 글미을 클래스파이한다. 간순히 url배끼는 것은 아님. 

#  이미지를 불러온다. url 주소가 등장한다. 
# 아 이게 그림이구나 확인. 
# 사이트를 확인하면, 대충 url해서 이미지를 오픈한다. 
```

Demo in Gradio

```python
def func (image): # 이미지 가정
  feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
  model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
# 그릠 파일이 제공된다면. ㄱ 정보를 바탕으로 함술ㄹ 제작한다. 
# 입축

  inputs = feature_extractor(images=image, return_tensors="pt")
  outputs = model(**inputs)
  logits = outputs.logits
  # model predicts one of the 1000 ImageNet classes
  predicted_class_idx = logits.argmax(-1).item()
  predicted_class = model.config.id2label[predicted_class_idx]
  return predicted_class
```

### [Fill-Mask](https://huggingface.co/tasks/fill-mask)

e.g. https://huggingface.co/bert-base-uncased \
How to use

```python
!pip install transformers
```

```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a [MASK] model.")

# 모델을 실형한다. 아디어나 방법은 쉽다. 
# 일단. enough를 정의 
```

Demo in Gradio

```python
import pandas as pd
def func (text):
  unmasker = pipeline('fill-mask', model='bert-base-uncased')
  result = unmasker(text)
  df = pd.DataFrame(result)
  return df

  # 함수를 만든다.
```

```python
examples = ["Hello I'm a [MASK] model.", "It is raining outside. I feel [MASK]."]
```

```python
gr.Interface(fn=func, inputs='text', outputs='dataframe', examples = examples).launch()

# 가능한 형용사를 만들어준다. 
```

### [Token classification](https://huggingface.co/tasks/token-classification)

e.g. https://huggingface.co/dslim/bert-base-NER \
How to use

```python
!pip install transformers

# 자연언어 이해 중 하나이다. 
# 어떤 자연언어에 대해서. 주어진 문장의 자유도가 너무 높을 수 있다. 
# 그 높은 자유화에서 뽑아낼 수 있어야 한다. 

# 주어진 문장에 대해서 그 중에서 어떤 명사는 더 정보적으로 더욱 중요하다. 
# 더 관심깊게 주목해야할 명사를 인지하는 기술이다. 
#named 엔티티. ㅌㅡ겆라한 것에 대해서 캐치한다.
```

recognize four types of entities: location (LOC), organizations (ORG), person (PER) and Miscellaneous (MISC).

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)

# 피처를 뽑고 전처릴 한다. 
# 토크나이저는 쪼개주는 역할을 한다. 단어 수준을 조갠다. 
# 그것을 받아놓으면 각각의 토크늗ㄹㅇㄴ 
# 딕셔너리가 2개로 남노. 
# 그것을 받아서 
# 최졸 리절트ㅡ르 받아준ㄴ다. 
```

Demo in Gradio

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
def func (text):
  tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
  model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
  nlp = pipeline("ner", model=model, tokenizer=tokenizer)
  result = nlp(text)
  df = pd.DataFrame(result)
  return df
```

```python
examples = ["My name is Wolfgang and I live in Berlin", "I will visit Seoul to see Chris"]
```

```python
gr.Interface(fn=func, inputs='text', outputs='dataframe', examples = examples).launch()
```

### [Sentence similarity](https://huggingface.co/tasks/sentence-similarity)

e.g. https://huggingface.co/sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2 \
How to use

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer, util
sentences = ["This is an example sentence", "it is one example writing"]

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
embeddings = model.encode(sentences)
print(embeddings)
```

```python
cosine_scores = util.pytorch_cos_sim(embeddings[0], embeddings[1])
cosine_scores

# 원점 과 평면을 만든다. 
```

Demo in Gradio

```python
def func (text):
  from sentence_transformers import SentenceTransformer, util
  model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
  embeddings = model.encode(text)
```

```python
examples = ["My name is Wolfgang and I live in Berlin", "I will visit Seoul to see Chris"]
```
