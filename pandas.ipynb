### Constructing a DataFrame using dict

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

# 딕트 안에 5개가 들어가 있다. 각 아이템들은 앞부분의 키와 밸류로 구분한다. 콜론으로 입력 
# 키부분에 - 칼럼의 이름-5개 5개 
# 키는 칼럼이 되고, 벨류는 리스트의 것들의 값이 된다. 
# 종 은 또한 그렇다., 

df = pd.DataFrame(data)
df
# 딕트를 넣어주면 데이터 프레임으로 만들어준다는 것임.일일이 적어주고 만들어줄 수 있다.  
# 딕트 변수를 넣어서 데이터 프레임이라는 함수를 통해 만들어줄 수 있다. 
```

### Constructing a DataFrame using pd.Series

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])
# 미리 만든 것이 아니라. series를 만들어준다. 


df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df

# 그 다음에 딕트를 만들어서 만들어준다. 
# 앞선 키는 문자열로 설정되어 있고, 그 키에 해당하는 밸류는 설정되어 있다. 그 밸류를 변수로 하여서 
```

### Changing column names

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df

# 특정정보만 엑세스하는 법을 배웠다. 수천개의 칼럼을  굳이 사용할 필요는 ㅇ벇을 듯. 일부의 뇽을 변경
# 또는 수정을 하는 방법. df 칼럼. 
# 수정할 수 있다. colums 사용시./ species 부분만 수정한다. 칼럼 명이 소문자로 변경된다. 
```

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']
# 행의 인덱스와 항렬의 인덱스를 넣어준다. 콜론이 들어가면 전부다 된다. 행에 대해서 전부다. 
# 칼럼 인덱스의 경우 'sepal_length' - 그 정보만 가져 온다.
# 앞에는 줄을 가져오고, 뒤에는 칼럼
```

```python
# Access one column (short-hand)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]

# 한 줄만 가지고 오고 싶을 때, 모든 로우에서 가져온다?
```

```python
# Access two rows
df.loc[[3, 0], :]
```

```python
# Access a range of rows
df.loc[2:5, :]

# 앞에는 줄을 뒤에는 칼럼을 가져 온다. 
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]

# 5보다 큰경우 가져옿ㄴ다. 이 경우에는? 5보 큳자. 
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4]
# 하나의 셀을 가져 오는 경우. 
# 한 칼럼을 통때로 가져와서 그 중에서 4번째에 해당하는 것을 가지고 온다. 
# 칼럼 억세싱하기. 몇번째 로우에 해당하는지. 하나의 셀도 엑세스 한다. 
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
# 리스트를 집어 넣어준다. 새로운 칼럼을 만들어준다. 
```
